```python
from __future__ import absolute_import, division, print_function

import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense, Conv1D, Flatten, SimpleRNN, LSTM
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
filename = "../Analysis/mfccFeatures_gender.csv"
```

```python
data = pd.read_csv(filename)
train, test = train_test_split(data, test_size=0.2)
train_data = train.iloc[:,0:20].copy()
train_label = train.iloc[:,20:].copy()
test_data = test.iloc[:,0:20].copy()
test_label = test.iloc[:,20:].copy()
print(data.shape)
print(train_data.shape)
print(train_label.shape)
print(test_data.shape)
print(test_label.shape)
```

```python
model = Sequential()
model.add(Dense(15,input_dim=20,activation='relu'))
model.add(Dense(10,activation='relu'))
model.add(Dense(1,activation='sigmoid'))

model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])

model.fit(train_data,train_label,epochs=150,batch_size=50)
```

```python
test_loss, test_acc = model.evaluate(test_data, test_label)
print("Test Accuracy: ", test_acc)
```

```python
#CNN Model
data = pd.read_csv(filename)
train, test = train_test_split(data, test_size=0.2)
train_data = np.array(train.iloc[:,0:50].copy())
train_label = np.array(train.iloc[:,50:].copy())
test_data = np.array(test.iloc[:,0:50].copy())
test_label = np.array(test.iloc[:,50:].copy())
print(data.shape)
print(train_data.shape)
print(train_label.shape)
print(test_data.shape)
print(test_label.shape)
```

```python
train_data[0].shape
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
model.summary()
```

```python
model.fit(train_data.reshape(len(train_data),50,1),toCategorical(train_label,8),epochs=150,batch_size=50)
```

```python
test_loss, test_acc = model.evaluate(test_data.reshape(len(test_data),50,1), toCategorical(test_label,8))
print("Test Accuracy: ", test_acc)
```
